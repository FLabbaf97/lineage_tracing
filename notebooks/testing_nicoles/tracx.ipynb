```python
import numpy as np

def calculate_idp_idr_idf1(ground_truth_matrix, predicted_matrix):
    true_positives = np.sum(np.logical_and(ground_truth_matrix, predicted_matrix))
    false_positives = np.sum(np.logical_and(np.logical_not(ground_truth_matrix), predicted_matrix))
    false_negatives = np.sum(np.logical_and(ground_truth_matrix, np.logical_not(predicted_matrix)))

    idp = true_positives / (true_positives + false_positives) if (true_positives + false_positives) != 0 else 0
    idr = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) != 0 else 0
    idf1 = 2 * (idp * idr) / (idp + idr) if (idp + idr) != 0 else 0

    return idp, idr, idf1
```

```python
# Example usage
ground_truth_matrix = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])  # Example ground truth assignment matrix
predicted_matrix = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 1]])  # Example predicted assignment matrix

idp, idr, idf1 = calculate_idp_idr_idf1(ground_truth_matrix, predicted_matrix)
print(f"IDP: {idp}, IDR: {idr}, IDF1: {idf1}")


ground_truth_matrix = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])  # Example ground truth assignment matrix
predicted_matrix = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])  # Example predicted assignment matrix

idp, idr, idf1 = calculate_idp_idr_idf1(ground_truth_matrix, predicted_matrix)
print(f"IDP: {idp}, IDR: {idr}, IDF1: {idf1}")
```

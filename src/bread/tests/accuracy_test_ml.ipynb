```python
# get two pd files and calculate metrics
import pandas as pd
from sklearn.metrics import accuracy_score

def calculate_metrics(lineage_GT, lineage_pred):
    # Merge the ground truth and prediction dataframes using the `time_index` and `bud_id` columns as keys
    lineage_merged = pd.merge(lineage_GT, lineage_pred, on=['time_index', 'bud_id'], suffixes=('_gt', '_pred'))
    
    # Compute the accuracy, precision, recall, and F1-score
    num_correct = (lineage_merged['# parent_id_gt'] == lineage_merged['# parent_id_pred']).sum()
    num_total = len(lineage_merged)
    accuracy = accuracy_score(lineage_merged['# parent_id_gt'], lineage_merged['# parent_id_pred'])
    
    tp = ((lineage_merged['# parent_id_gt'] == lineage_merged['# parent_id_pred']) & (lineage_merged['# parent_id_gt'] != -1)).sum()
    fp = ((lineage_merged['# parent_id_gt'] != lineage_merged['# parent_id_pred']) & (lineage_merged['# parent_id_gt'] == -1)).sum()
    fn = ((lineage_merged['# parent_id_gt'] != lineage_merged['# parent_id_pred']) & (lineage_merged['# parent_id_pred'] == -1)).sum()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * precision * recall / (precision + recall)
    print('Accuracy: ' + str(accuracy) + '\n' + 'Precision: ' + str(precision) + '\n' + 'Recall: ' + str(recall) + '\n' + 'F1-score: ' + str(f1_score))
    return accuracy, precision, recall, f1_score
    
```
